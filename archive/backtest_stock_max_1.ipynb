```python
sub_list = [
    # 科创100
    "SH.688001",
    "SH.688002",
    "SH.688006",
    "SH.688007",
    "SH.688016",
    "SH.688017",
    "SH.688018",
    "SH.688019",
    "SH.688023",
    "SH.688029",
    "SH.688033",
    "SH.688037",
    "SH.688050",
    "SH.688055",
    "SH.688072",
    "SH.688083",
    "SH.688088",
    "SH.688097",
    "SH.688100",
    "SH.688105",
    "SH.688106",
    "SH.688107",
    "SH.688110",
    "SH.688116",
    "SH.688120",
    "SH.688123",
    "SH.688127",
    "SH.688128",
    "SH.688131",
    "SH.688133",
    "SH.688139",
    "SH.688153",
    "SH.688161",
    "SH.688166",
    "SH.688183",
    "SH.688185",
    "SH.688196",
    "SH.688198",
    "SH.688200",
    "SH.688202",
    "SH.688208",
    "SH.688213",
    "SH.688232",
    "SH.688233",
    "SH.688248",
    "SH.688261",
    "SH.688266",
    "SH.688268",
    "SH.688275",
    "SH.688276",
    "SH.688278",
    "SH.688281",
    "SH.688289",
    "SH.688298",
    "SH.688311",
    "SH.688315",
    "SH.688318",
    "SH.688321",
    "SH.688326",
    "SH.688327",
    "SH.688331",
    "SH.688333",
    "SH.688339",
    "SH.688356",
    "SH.688366",
    "SH.688368",
    "SH.688383",
    "SH.688387",
    "SH.688388",
    "SH.688389",
    "SH.688408",
    "SH.688409",
    "SH.688425",
    "SH.688439",
    "SH.688499",
    "SH.688508",
    "SH.688516",
    "SH.688518",
    "SH.688520",
    "SH.688526",
    "SH.688556",
    "SH.688559",
    "SH.688567",
    "SH.688568",
    "SH.688575",
    "SH.688578",
    "SH.688586",
    "SH.688598",
    "SH.688608",
    "SH.688617",
    "SH.688658",
    "SH.688680",
    "SH.688686",
    "SH.688696",
    "SH.688700",
    "SH.688707",
    "SH.688739",
    "SH.688778",
    "SH.688789",
    "SH.688798",
]

sub_list_1 = [
    "SZ.300212",
    "SZ.300002",
    "SZ.300017",
    "SZ.300442",
    "SZ.300476",
    "SZ.300573",
    "SZ.300765",
    "SZ.301301",
    "SZ.300114",
    "SZ.300394",
    "SZ.300502",
    "SZ.300567",
    "SZ.301267",
    # "SZ.301358",
    "SZ.300633",
    "SZ.300666",
    "SZ.300803",
    "SZ.301207",
    "SZ.301236",
    "SZ.301269",
    "SZ.301308",
    "SZ.300054",
    # "SZ.300118",
    # "SZ.300034",
    "SZ.300395",
    "SZ.300487",
    "SZ.301035",
    "SZ.301029",
    "SZ.300776",
    "SZ.300604",
    "SZ.300373",
    "SZ.300390",
    "SZ.300568",
    "SZ.300769",
    "SZ.300861",
    "SZ.300919",
    "SZ.300957",
    "SZ.300979",
    "SZ.300763",
    "SZ.300751",
    "SZ.300866",
    "SZ.300896",
    "SZ.300888",
    "SZ.300832",
    "SZ.300037",
    "SZ.300999",
    "SZ.300223",
    "SZ.300775",
    "SZ.300782",
    "SZ.300661",
    "SZ.300682",
    "SZ.300759",
    "SZ.300595",
    "SZ.300724",
    "SZ.300748",
    "SZ.300357",
    "SZ.300012",
    "SZ.300454",
    "SZ.300529",
    "SZ.300760",
    "SZ.300750",
    "SZ.300601",
    "SZ.300285",
    "SZ.300308",
    "SZ.300413",
    "SZ.300474",
    "SZ.300699",
    "SZ.300628",
    "SZ.300676",
    "SZ.300316",
    "SZ.300558",
    "SZ.300450",
    "SZ.300073",
    "SZ.300294",
    "SZ.300296",
    "SZ.300009",
    "SZ.300014",
    "SZ.300496",
    "SZ.300383",
    "SZ.300418",
    "SZ.300498",
    "SZ.300136",
    "SZ.300408",
    "SZ.300433",
    "SZ.300033",
    "SZ.300253",
    "SZ.300207",
    "SZ.300347",
    "SZ.300251",
    "SZ.300274",
    "SZ.300144",
    "SZ.300146",
    "SZ.300122",
    "SZ.300124",
    "SZ.300142",
    "SZ.300001",
    "SZ.300003",
    "SZ.300015",
    "SZ.300059",
    "SZ.300070",
]

sub_list_2 = [
    # 科创50
    "SH.688008",
    "SH.688009",
    "SH.688012",
    "SH.688036",
    "SH.688041",
    "SH.688047",
    "SH.688065",
    # "SH.688072",
    "SH.688082",
    "SH.688099",
    "SH.688111",
    "SH.688114",
    "SH.688120",
    "SH.688122",
    "SH.688126",
    "SH.688169",
    "SH.688180",
    "SH.688187",
    "SH.688188",
    "SH.688220",
    "SH.688223",
    "SH.688234",
    # "SH.688249",
    "SH.688256",
    "SH.688271",
    "SH.688295",
    "SH.688297",
    "SH.688301",
    "SH.688303",
    "SH.688349",
    "SH.688363",
    "SH.688375",
    "SH.688385",
    "SH.688396",
    # "SH.688469",
    # "SH.688472",
    "SH.688475",
    # "SH.688506",
    "SH.688521",
    "SH.688525",
    "SH.688538",
    "SH.688561",
    # "SH.688563",
    "SH.688599",
    "SH.688617",
    "SH.688728",
    "SH.688777",
    "SH.688819",
    "SH.688981",
    "SH.689009",
]

import polars as pl
import os
import time
import numpy as np

from dataclasses import dataclass
from src.futuapi.futu_api import FutuApi
from src.calculator.mathmatics import add_mean, add_boll

# 设置 Polars 显示的最大行数和列数
# pl.Config.set_tbl_width(10000)  # 设置表格的最大宽度，单位为字符
pl.Config.set_tbl_cols(100)  # 设置显示的最大列数
pl.Config.set_tbl_rows(100)  # 设置显示的最大行数


def rolling_percentile(arr, p):
    return np.percentile(arr, p)


rolling_day = 10


def parse_df_time(df):
    df = df.with_columns(
        pl.col("time_key").str.strptime(pl.Datetime, format="%Y-%m-%d %H:%M:%S"),
        pl.col("close").cast(pl.Float64).alias("price"),
    )
    df = df.with_columns(
        [
            pl.col("time_key").dt.date().cast(pl.Utf8).alias("date"),
            pl.col("time_key").dt.time().cast(pl.Utf8).alias("time"),
        ]
    )
    return df


def trans(df: pl.DataFrame):
    df = parse_df_time(df)
    df_lh = df.group_by("date").agg(
        [
            pl.col("high").max().alias("high"),
            pl.col("low").min().alias("low"),
        ]
    )
    df_check_vol = (
        df.filter(pl.col("time") <= "14:50:00")
        .group_by("date")
        .agg([pl.col("volume").sum().alias("check_volume")])
        .sort("date")
    )
    df_close = (
        df.filter(pl.col("time") == "15:00:00")
        .with_columns(pl.col("close").cast(pl.Float64).alias("close"))
        .select("date", "close", "volume")
    )
    df_check_price = (
        df.filter(pl.col("time") == "14:50:00")
        .with_columns(pl.col("price").cast(pl.Float64).alias("check_price"))
        .select("date", "check_price")
    )
    df = (
        (df_check_price.join(df_check_vol, on=["date"]))
        .select(["date", "check_volume", "check_price"])
        .join(df_close, on=["date"])
        .join(df_lh, on=["date"])
    )
    df = df.sort("date")
    # print(index_df)
    df = df.with_columns(
        pl.col("check_price")
        # .shift(1)
        .rolling_max(window_size=25).alias("max_25")
    )
    df = df.with_columns(
        pl.col("check_volume")
        # .shift(1)
        .rolling_mean(window_size=10)
        .cast(pl.Int64)
        .alias("avg_vol_10")
    )

    # av_price
    # df = df.with_columns(pl.col("close").rolling_mean(20).alias(f"ma_20"))
    df = df.with_columns(
        [
            df["close"]
            .rolling_map(
                function=lambda x: rolling_percentile(x, 30),
                window_size=rolling_day,
                weights=None,
                min_periods=1,  # 至少 1 个数据点，否则返回 NaN
            )
            .alias("P30"),
            df["close"]
            .rolling_map(
                function=lambda x: rolling_percentile(x, 70),
                window_size=rolling_day,
                weights=None,
                min_periods=1,
            )
            .alias("P70"),
        ]
    )
    df = df.with_columns((df["P70"] - df["P30"]).alias("P70_minus_P30"))
    df = df.with_columns(
        pl.col("close").rolling_mean(rolling_day).alias(f"ma_{rolling_day}")
    )
    df = df.with_columns(
        (df["P70_minus_P30"] / df[f"ma_{rolling_day}"] * 100).alias("av_rate")
    ).drop(["P70_minus_P30", "P70", "P30", f"ma_{rolling_day}"])

    # av_volume
    df = df.with_columns(
        [
            df["volume"]
            .rolling_map(
                function=lambda x: rolling_percentile(x, 30),
                window_size=rolling_day,
                weights=None,
                min_periods=1,  # 至少 1 个数据点，否则返回 NaN
            )
            .alias("P30"),
            df["volume"]
            .rolling_map(
                function=lambda x: rolling_percentile(x, 70),
                window_size=rolling_day,
                weights=None,
                min_periods=1,
            )
            .alias("P70"),
        ]
    )
    df = df.with_columns((df["P70"] - df["P30"]).alias("P70_minus_P30"))
    df = df.with_columns(
        pl.col("volume").rolling_mean(rolling_day).alias(f"vol_ma_{rolling_day}")
    )
    df = df.with_columns(
        (df["P70_minus_P30"] / df[f"vol_ma_{rolling_day}"] * 100).alias("av_rate_vol")
    ).drop(["P70_minus_P30", "P70", "P30", f"vol_ma_{rolling_day}"])

    df = df.with_columns(pl.col("close").rolling_mean(20).alias(f"ma_20"))
    df = df.with_columns(pl.col("close").rolling_mean(30).alias(f"ma_30"))
    # df = df.with_columns(pl.col("price").rolling_mean(120).alias(f"ma_120"))

    # df = df.with_columns(
    #     pl.col("price").rolling_min(window_size=5, min_periods=1).alias("min_5")
    # )
    # df = df.with_columns([(pl.col("price") - pl.col("price").shift(5)).alias("del_10")])

    # print(df)
    # time.sleep(10)

    return df


@dataclass
class SecRecord:
    date: str
    code: str
    check_price: float
    check_volume: int
    max_25: float
    avg_vol_10: int
    av_rate: float
    high: float
    low: float
    av_rate_vol: int
    close: float
    ma_20: float
    ma_30: float


class MarketDataMap:
    def __init__(self):
        self.raw_data = {}
        self.start_date = "2023-01-01"
        self.end_date = "2024-12-06"

    def init_data(self):
        api = FutuApi()
        try:
            for sec_id in [
                # "SZ.300628",
                *sub_list_1,
                *sub_list,
                *sub_list_2,
            ]:
                print(sec_id)
                df = api.get_minute_kline(sec_id, self.start_date, self.end_date)
                df = trans(df)
                # print(df)
                for row in df.iter_rows(named=True):
                    date = row["date"]
                    if date not in self.raw_data:
                        self.raw_data[date] = []
                    # print(row)
                    record = SecRecord(
                        code=sec_id,
                        date=row["date"],
                        check_price=row["check_price"],
                        check_volume=row["check_volume"],
                        max_25=row["max_25"],
                        avg_vol_10=row["avg_vol_10"],
                        av_rate=row["av_rate"],
                        low=row["low"],
                        high=row["high"],
                        av_rate_vol=row["av_rate_vol"],
                        close=row["close"],
                        ma_20=row["ma_20"],
                        ma_30=row["ma_30"],
                    )
                    self.raw_data[date].append(record)
        finally:
            api.close()

        # for date, data in self.raw_data.items():
        #     print(date, data)
        #     time.sleep(1)


market_data_map = MarketDataMap()
market_data_map.init_data()
```

```python
import math
import datetime
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator


def truncate_float(number, t):
    """
    截断浮点数到指定的小数位数，不进行四舍五入。

    :param number: 要截断的浮点数
    :param decimals: 小数位数
    :return: 截断后的浮点数
    """
    factor = 10.0**t
    truncated = math.trunc(number * factor) / factor
    return truncated


def show_rate(x: list, y: list):
    ax = plt.gca()  # 获取当前轴对象
    ax.xaxis.set_major_locator(
        MaxNLocator(integer=True, prune="both", nbins=6)
    )  # 自动选择6个关键点显示
    plt.plot(x, y, linestyle="-", linewidth=1, color="b", label="y = raw")
    # plt.plot(x, py, linestyle="-", linewidth=1, color="r", label="py = main")
    plt.title(f"str_max_fund_1")
    plt.xlabel("date")
    plt.ylabel("rate")
    plt.xticks(rotation=45, fontsize=10)
    plt.tight_layout()
    plt.grid(True, linestyle="--", alpha=0.5)
    # 显示图例
    plt.legend()

    # 显示图形
    plt.show()


class SecSignal:
    def __init__(self):
        self.position_status = "no_postion"
        self.price = 0.0
        self.rate = 0.0
        self.max_price = 0.0
        self.max_25 = None
        self.avg_vol_10 = None
        self.av_rate = 0.0
        self.av_rate_vol = 0.0
        self.close = 0.0
        self.ma_20 = 0.0
        self.ma_30 = 0.0
        self.last_check_volume = 0

    def load(self, record: SecRecord):
        if record.max_25 == None or record.ma_30 == None:
            return "None"
        elif self.max_25 == None:
            self.max_25 = record.max_25
            self.avg_vol_10 = record.avg_vol_10
            self.av_rate = record.av_rate
            self.av_rate_vol = record.av_rate_vol
            self.close = record.close
            self.ma_20 = record.ma_20
            self.ma_30 = record.ma_30
            self.last_check_volume = record.check_volume
            return "None"

        result = "None"
        if (
            self.position_status == "no_postion"
            and record.check_price >= self.max_25
            and record.check_volume > self.avg_vol_10 * 1.5
            and self.av_rate < 3.0
            and self.close * 1.195 >= record.check_price
            # and self.close * 1.05 < record.check_price
            and self.ma_20 >= self.ma_30
        ):
            self.price = record.check_price
            self.on_position = True
            self.max_price = record.check_price
            self.position_status = "1"
            result = "Buy"
        elif self.position_status == "1":
            if record.check_price < self.close:
                result = "Buy"
                self.position_status = "2"
            elif (
                record.check_price > self.close * 1.05
                and record.check_volume > self.last_check_volume
            ):
                self.position_status = "1.1"
                result = "None"
            else:
                result = "Sale"
                self.position_status = "no_position"
        elif self.position_status == "2":
            if record.check_price < self.close:
                result = "Buy"
                self.position_status = "3"
            else:
                result = "Sale"
                self.position_status = "no_position"
        elif self.position_status == "3" and record.check_price > self.close:
            result = "Sale"
            self.position_status = "no_position"
        elif self.position_status == "1.1":
            if (
                record.check_price > self.close * 1.05
                and record.check_volume > self.last_check_volume
            ):
                pass
            else:
                result = "Sale"
                self.position_status = "no_position"

        self.ma_20 = record.ma_20
        self.ma_30 = record.ma_30
        self.last_check_volume = record.check_volume
        self.max_25 = record.max_25
        self.avg_vol_10 = record.avg_vol_10
        self.av_rate = record.av_rate
        self.av_rate_vol = record.av_rate_vol
        self.close = record.close
        self.max_price = max(self.max_price, record.check_price)
        return result

    def check(self, record: SecRecord):
        result = "None"
        print(record, self.avg_vol_10, self.max_25)
        if self.on_position:
            if (
                record.check_index_price <= self.max_25 * 0.95
                or record.check_volume < self.avg_vol_10
            ):
                result = "Sale"
        else:
            if (
                record.check_index_price >= self.max_25
                and record.check_volume > self.avg_vol_10 * 1.2
            ):
                result = "Buy"
        return result


class FundAccount:
    def __init__(self):
        self.init = 100000
        self.cash = 100000
        self.position = {}
        self.fee = 0.001
        # self.buys = []
        # self.sales = []
        # self.has_position = 0
        # self.no_position = 0
        self.max_val = 0.0
        self.max_drawdone = 0.0
        self.rate = 0.0

        self.max_val = 0.0
        self.max_drawdone = 0.0

        self.has_position_days = 0
        self.days = 0

    def val(self, records: list):
        result = 0.0
        result += self.cash
        result += sum(
            [
                self.position[record.code][0] * record.check_price
                for record in records
                if record.code in self.position
            ]
        )
        return result

    def buy_record(self, deno: int, record: SecRecord):
        t_cash = self.cash / deno
        t_price = record.check_price * (1 + self.fee)
        cnt = truncate_float(t_cash / t_price, -2)
        if record.code in self.position:
            self.position[record.code][0] += cnt
            self.position[record.code][1] += 1
        else:
            self.position[record.code] = [cnt, 1]
        self.cash -= cnt * t_price

    def buy(self, buy_list: list):
        i = 0
        for deno in range(
            10 - sum([ele[1] for code, ele in self.position.items()]), 0, -1
        ):
            if len(buy_list) <= i:
                break
            self.buy_record(deno, buy_list[i][0])
            i += 1

    def sale_record(self, record: SecRecord):
        if record.code not in self.position:
            # print(f"err: not find record postion for code {record.code}")
            return
        t_price = record.check_price * (1.0 - self.fee)
        self.cash += self.position[record.code][0] * t_price
        del self.position[record.code]

    def sale(self, sale_list: list):
        for [record, rate] in sale_list:
            self.sale_record(record)

    def print_meta(self, records: list):
        val = self.val(records)
        self.days += 1
        if len(self.position) != 0:
            self.has_position_days += 1
        self.max_val = max(self.max_val, val)
        self.max_drawdone = min(self.max_drawdone, val * 100 / self.max_val - 100.0)
        print(
            f"max_val: {self.max_val}, max_drawdone: {self.max_drawdone}, rate: {val * 100 / self.init - 100.0}, has position day: {self.has_position_days}, has position rate: {self.has_position_days * 100 / self.days}"
        )


fund_account = FundAccount()


class StrStockMax1:
    def __init__(self):
        self.sub_list = [
            *sub_list,
            *sub_list_1,
            *sub_list_2,
        ]
        self.date = datetime.datetime.today().strftime("%Y-%m-%d")
        self.signals = dict([(code, SecSignal()) for code in self.sub_list])

    def load_records(self):
        x = []
        y = []

        for date, records in market_data_map.raw_data.items():
            buy_list = []
            sale_list = []
            for record in records:
                # if record.code == "SZ.300442":
                #     print(f"record: {record}")
                sig = self.signals[record.code].load(record)
                if sig == "Sale":
                    sale_list.append([record, self.signals[record.code].rate])
                elif sig == "Buy":
                    buy_list.append([record])

            fund_account.sale(sale_list)
            fund_account.buy(buy_list)
            val = fund_account.val(records)
            # fund_account.print_meta(records)

            print(
                date,
                "sale: ",
                [[ele[0].code, ele[1]] for ele in sale_list],
                "buy: ",
                [[ele[0].code] for ele in buy_list],
                "val: ",
                val,
            )
            x.append(date)
            y.append(val * 100 / 100000 - 100.0)

            # if date == "2023-03-13":
            #     break

        show_rate(x, y)


strategy = StrStockMax1()
strategy.load_records()
```
